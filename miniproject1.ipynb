```python
import pandas as pd

# Step 1: Create dataset automatically
data = {
    "v1": ["ham","spam","ham","spam","ham","spam","ham","spam","ham","spam"],
    "v2": [
        "Hey how are you?",
        "Win money now!!!",
        "Let's study together",
        "Free iPhone offer click now",
        "Are you coming to class?",
        "Earn cash fast!!!",
        "Call me later",
        "Limited offer buy now",
        "See you tomorrow",
        "You won a lottery claim now"
    ]
}

df = pd.DataFrame(data)

# Save as CSV
df.to_csv("spam.csv", index=False)

print("✅ spam.csv file created!")

# Step 2: Train model
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

X = df['v2']
y = df['v1']

cv = CountVectorizer()
X = cv.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = MultinomialNB()
model.fit(X_train, y_train)

# Step 3: Test
msg = ["Free money waiting for you"]
prediction = model.predict(cv.transform(msg))

print("Message:", msg[0])
print("Prediction:", prediction[0])

# Step 4: Accuracy
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```
