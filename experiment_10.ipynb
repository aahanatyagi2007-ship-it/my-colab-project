```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = {
    'Name': ['Aman', 'Riya', 'Karan', 'Sneha', 'Rahul'],
    'Age': [18, 19, 18, 20, 19],
    'Marks': [85, 90, 78, 88, 92],
    'Height': [5.7, 5.4, 5.8, 5.5, 5.9],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Chandigarh', 'Mumbai']
}

df = pd.DataFrame(data)

print("Dataset:\n")
print(df)




plt.figure(figsize=(8,5))
plt.plot(df['Name'], df['Marks'], marker='o')
plt.title("Line Plot of Marks")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.grid(True)
plt.show()


plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['Marks'])
plt.title("Bar Chart of Marks")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


plt.figure(figsize=(8,5))
plt.hist(df['Marks'], bins=5, edgecolor='black')
plt.title("Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()


plt.figure(figsize=(8,5))
plt.scatter(df['Age'], df['Marks'])
plt.title("Scatter Plot of Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()

plt.figure(figsize=(6,6))
plt.pie(df['Marks'], labels=df['Name'], autopct='%1.1f%%')
plt.title("Pie Chart of Marks")
plt.show()




sns.boxplot(y=df['Marks'])
plt.title("Box Plot of Marks")
plt.show()


plt.figure(figsize=(8,5))
sns.countplot(x='City', data=df)
plt.title("Count Plot of Students by City")
plt.show()

plt.figure(figsize=(8,5))
sns.heatmap(df[['Age', 'Marks', 'Height']].corr(), annot=True, cmap='coolwarm')
plt.title("Heatmap of Correlation Matrix")
plt.show()


sns.pairplot(df[['Age', 'Marks', 'Height']])
plt.show()
```
